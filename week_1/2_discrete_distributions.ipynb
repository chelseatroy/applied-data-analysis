## Common Discrete Distributions

We've been building up intuition with dice and coins. Now let's look at some named distributions that show up constantly in data analysis and machine learning. Each one has a natural story (a type of situation it describes) and a set of parameters that let you tune it.

### Bernoulli Distribution

The simplest distribution: a single yes/no trial. Did the coin land heads? Did the user click the ad? Did the email get marked as spam?

There's exactly one parameter: **p**, the probability of "success" (however you define it). The probability of failure is 1 - p.

If p = 0.5, it's a fair coin. If p = 0.3, success happens 30% of the time.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed

# Fix the random seed so results are reproducible
rng = np.random.default_rng(42)

# Nice plot defaults
plt.rcParams['figure.dpi'] = 100
sns.set_theme(style='whitegrid')
```

```python
def plot_bernoulli(p=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.5, continuous_update=False, description='p')):
    fig, ax = plt.subplots(figsize=(8, 4))
    outcomes_b = [0, 1]
    probs_b = [1 - p, p]
    ax.bar(outcomes_b, probs_b, color=['coral', 'steelblue'], edgecolor='white', width=0.4)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['Failure (0)', 'Success (1)'])
    ax.set_ylabel('Probability')
    ax.set_xlabel('Outcome')
    ax.set_title(f'Bernoulli Distribution (p={p:.2f})')
    ax.set_ylim(0, 1.1)
    for x, prob in zip(outcomes_b, probs_b):
        ax.text(x, prob + 0.02, f'{prob:.2f}', ha='center', fontsize=11)
    plt.tight_layout()
    plt.show()

interact(plot_bernoulli)
```

### Binomial Distribution

What if instead of one coin flip, you flip a coin 10 times? The **Binomial distribution** counts the number of successes in **n** independent Bernoulli trials, each with probability **p** of success.

Parameters:
- **n** = number of trials
- **p** = probability of success on each trial

Example: out of 20 patients who receive a treatment with a 60% success rate, how many recover? That's Binomial(n=20, p=0.6).

As n gets large, the binomial distribution starts to look bell-shaped.

```python
def plot_binomial(
    n=widgets.IntSlider(min=1, max=50, step=1, value=10, continuous_update=False, description='n'),
    p=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.5, continuous_update=False, description='p')
):
    fig, ax = plt.subplots(figsize=(8, 4))
    k = np.arange(0, n + 1)
    pmf = stats.binom.pmf(k, n, p)
    ax.bar(k, pmf, color='steelblue', edgecolor='white', width=0.7)
    ax.axvline(x=n * p, color='red', linestyle='--', linewidth=2, label=f'Mean = {n*p:.2f}')
    ax.set_xlabel('Number of Successes')
    ax.set_ylabel('Probability')
    ax.set_title(f'Binomial Distribution (n={n}, p={p:.2f})')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

interact(plot_binomial)
```

### Geometric Distribution

Same setup as Binomial — independent trials, probability p of success each time — but now we ask a different question: **how many trials does it take to get the first success?**

Example: you're cold-calling potential customers who each have a 20% chance of saying yes. How many calls until you make your first sale?

The shape is always right-skewed (a long tail to the right): success on the first try is most likely, and the probability of needing more and more tries keeps decreasing.

```python
def plot_geometric(p=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.3, continuous_update=False, description='p')):
    fig, ax = plt.subplots(figsize=(8, 4))
    max_k = min(30, int(10 / p))
    k = np.arange(1, max_k + 1)
    pmf = stats.geom.pmf(k, p)
    ax.bar(k, pmf, color='coral', edgecolor='white', width=0.7)
    mean_val = 1 / p
    ax.axvline(x=mean_val, color='navy', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.2f}')
    ax.set_xlabel('Number of Trials Until First Success')
    ax.set_ylabel('Probability')
    ax.set_title(f'Geometric Distribution (p={p:.2f})')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

interact(plot_geometric)
```

### Poisson Distribution

The Poisson distribution models how many times something happens in a fixed window of time (or space), when events occur at a known average rate and each event is independent of the others.

Examples:
- How many emails arrive in an hour?
- How many customers enter a store in a day?
- How many bugs are in a 1000-line codebase?

The only parameter is **lambda** (often written λ), which is the average number of events in the window. The mean and variance of a Poisson distribution are both equal to lambda — a handy fact to remember.

```python
def plot_poisson(lambda_=widgets.FloatSlider(min=0.5, max=20.0, step=0.5, value=5.0, continuous_update=False, description='lambda')):
    fig, ax = plt.subplots(figsize=(8, 4))
    max_k = max(20, int(lambda_ * 2.5))
    k = np.arange(0, max_k + 1)
    pmf = stats.poisson.pmf(k, lambda_)
    ax.bar(k, pmf, color='seagreen', edgecolor='white', width=0.7)
    ax.axvline(x=lambda_, color='darkred', linestyle='--', linewidth=2, label=f'Mean = {lambda_:.1f}')
    ax.set_xlabel('Number of Events')
    ax.set_ylabel('Probability')
    ax.set_title(f'Poisson Distribution (lambda={lambda_:.1f})')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

interact(plot_poisson)
```

### Negative Binomial Distribution

The Geometric distribution asks: how many trials until the first success? The **Negative Binomial distribution** extends that idea: how many trials until we get **r successes**?

Parameters:
- **r** = the number of successes you are waiting for
- **p** = the probability of success on each trial

Example: you want to win the lottery twice. Each ticket has a small probability p of winning. How many tickets do you need to buy before you win twice? That count follows a Negative Binomial distribution.

When r = 1, the Negative Binomial is exactly the Geometric distribution. As r increases, you are waiting for more successes, so the total number of trials needed grows — and the distribution shifts to the right and spreads out.

```python
def plot_negative_binomial(
    r=widgets.IntSlider(min=1, max=20, step=1, value=3, continuous_update=False, description='r'),
    p=widgets.FloatSlider(min=0.01, max=0.99, step=0.01, value=0.3, continuous_update=False, description='p')
):
    fig, ax = plt.subplots(figsize=(8, 4))
    max_k = int(stats.nbinom.ppf(0.999, r, p)) + 1
    k = np.arange(0, max_k + 1)
    pmf = stats.nbinom.pmf(k, r, p)
    ax.bar(k, pmf, color='mediumpurple', edgecolor='white', width=0.7)
    mean_val = r * (1 - p) / p
    ax.axvline(x=mean_val, color='darkred', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.2f}')
    ax.set_xlabel('Number of Failures Before r-th Success')
    ax.set_ylabel('Probability')
    ax.set_title(f'Negative Binomial Distribution (r={r}, p={p:.2f})')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

interact(plot_negative_binomial)
```

### Hypergeometric Distribution

The **Hypergeometric distribution** is used when you are sampling **without replacement** from a finite pool. Unlike the Binomial distribution — where each trial is independent and the probabilities stay the same — here every draw changes what is left in the pool, so the trials are not independent.

Parameters:
- **N** = total number of items in the pool
- **K** = number of "desirable" items in the pool
- **n** = how many items you draw

Example: a lottery has 53 numbered balls, and 6 of them will be drawn as winning numbers. You pick 6 numbers on your ticket. How many of your picks match the winning numbers? That count follows a Hypergeometric distribution, because the winning numbers are drawn without replacement — the same ball cannot come up twice.

The Hypergeometric is similar to the Binomial, but corrects for the fact that the pool is finite and shrinks with each draw.

```python
def plot_hypergeometric(
    N=widgets.IntSlider(min=10, max=100, step=1, value=53, continuous_update=False, description='N (pool)'),
    K=widgets.IntSlider(min=1, max=100, step=1, value=6, continuous_update=False, description='K (desirable)'),
    n=widgets.IntSlider(min=1, max=100, step=1, value=6, continuous_update=False, description='n (drawn)')
):
    # Clamp K and n so they cannot exceed N
    K = min(K, N)
    n = min(n, N)

    fig, ax = plt.subplots(figsize=(8, 4))
    # Possible matches range from max(0, n+K-N) to min(n, K)
    low = max(0, n + K - N)
    high = min(n, K)
    k = np.arange(low, high + 1)
    pmf = stats.hypergeom.pmf(k, N, K, n)
    ax.bar(k, pmf, color='teal', edgecolor='white', width=0.7)
    mean_val = n * K / N
    ax.axvline(x=mean_val, color='darkred', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.2f}')
    ax.set_xlabel('Number of Desirable Items Drawn')
    ax.set_ylabel('Probability')
    ax.set_title(f'Hypergeometric Distribution (N={N}, K={K}, n={n})')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()

interact(plot_hypergeometric)
```
