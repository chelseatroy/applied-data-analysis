# What is Statistical Learning?

Every statistical learning problem starts with the same basic setup: we have some input data (called predictors, features, or independent variables) and some output we want to predict or understand (called the response or dependent variable). The goal is to find a function that maps from inputs to outputs — or at least a good approximation of one.

For example, we might have advertising spending on TV, radio, and the internet as inputs, and sales figures as the output. We can't directly control sales, but we can control advertising budgets, so understanding how the two are related is genuinely useful.

### Two reasons to build a model

There are two main reasons you might want to fit a model:

**Prediction**: Sometimes you just need to know what Y will be given some X, and you don't particularly care *why*. You treat the model as a black box. The model's only job is to be accurate.

**Analysis**: Other times, you want to understand *how* X affects Y. Which predictors matter? How much does Y change when X goes up by one unit? Is the relationship linear or more complicated? For analysis, the inner workings of the model matter.

Many real problems involve a mix of both.

### Parametric vs. non-parametric methods

**Parametric methods** start by assuming a shape for the function — for example, that it's a straight line. Then they use the data to estimate the parameters of that shape (like the slope and intercept of a line). The advantage: you only need to estimate a small number of parameters. The disadvantage: if the true relationship isn't that shape, your model will be off.

**Non-parametric methods** make no assumption about the shape of the function. Instead, they try to find an estimate that follows the data closely. The advantage: they can fit a much wider variety of relationships. The disadvantage: they need a lot more data to do it well, and they can be harder to interpret.

### The bias-variance tradeoff

This is one of the most important ideas in all of machine learning. Here's the core insight:

- A **simple model** (like a straight line) is easy to interpret and doesn't change much if you swap in different training data — but it might be systematically wrong (high bias, low variance).
- A **complex model** (like a wiggly curve that passes through every training point) fits the training data perfectly — but it's so tuned to the noise in that particular dataset that it fails badly on new data (low bias, high variance).

**Overfitting** is what happens when a model is too complex: it memorizes the training data instead of learning the underlying pattern. The model performs great on training data but poorly on new, unseen data.

The goal is to find the sweet spot: complex enough to capture real patterns, simple enough not to chase noise.

### Training error vs. test error

We measure how well a model works using error. But there's a critical distinction:

- **Training error**: how well the model fits the data it was trained on. This always gets better as the model gets more complex.
- **Test error**: how well the model predicts *new* data it hasn't seen. This is what actually matters.

A good model minimizes test error, not training error. As model complexity increases, training error keeps falling, but test error eventually starts rising again — that's the overfitting zone. We want to stop just before that happens.
